```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
from pydub import AudioSegment
from pydub.playback import play
```

```python
data, rate = librosa.load("207124-3-2-0.wav", sr=None)

def show_data(data):
    plt.figure()
    plt.title("data")
    plt.plot(data)
    plt.show()

show_data(data)
ipd.Audio(data, rate=44100)
```

```python
# extract mfcc file

data = librosa.feature.mfcc(data)
print('before transform ', data[0][:20])
```

```python
# Reconstruct the audio file

wav = librosa.feature.inverse.mfcc_to_audio(data)
show_data(wav)
ipd.Audio(wav, rate=44100)
```
